# Example 1a: Générer une sequence de P points consitent de N groupes centrés autour de $(x_i, y_i)$ ou $i$ identifie chaque groupe avec une variance gaussienne spaciale $\sigma_i$.

```python
# definire une nouvelle fonction: "def nom_de_la_fonction(arguments_positionel, argument_nommés=default):"
## les classes des arguments non pas a etre definis explictement. Un usage incorrect result en une erreur.

def some_sequence(P, positions, jitter, return_group=False):
    '''On peut écrire une description qui va ce montrer en tooltip entre triple guilemets'''
    
    # test si les arguments sont consistent
    assert (len(positions)==len(jitter)), 'mismatch'
    
    group = np.random.randint(len(positions), size=(P,))
    A = positions[group] + \
        np.random.normal(0,1,size=(P,)+positions.shape[1:]) \
        * jitter[group].reshape((P,)+(1,)*len(positions.shape[1:]))

    if return_group:
        return A, group
    else:
        return A
```

```python
A = some_sequence(2000, np.random.normal(0,1,size=(4)), np.ones(4)*.1)
```

```python
fig = plt.figure(figsize=(12,4))
_=plt.hist(A, bins=200, range=(-3,3))
plt.xlabel('X')
plt.ylabel('Count')
plt.xlim([-3,3])
```

# Example 1b: Represention graphique de la sequence precedente pour P=1000, N=4 et les parametres sont pris au hazard dans une distribution normale $\mathcal{N}(0,1)$ et $\sigma_i=1 \forall i$

```python
A, g = some_sequence(2000, np.random.normal(0,1,size=(4,2)).astype(np.float32), np.ones(4)*.3, return_group=True)
```

# Example 2: Optimization de fonction lineaire

```python
from scipy.stats import linregress
```

```python
slope, intercept, r, p, se = linregress(x=A[:, 0], y=A[:,1])
print (slope, intercept, r, p, se)
```

```python
from scipy.optimize import curve_fit
```

```python
def f(x, a, b):
    return a*x+b
```

```python
popt, pcov = curve_fit(f, A[:, 0], A[:,1])
perr = np.sqrt(np.diag(pcov))
print (popt, '+/-', perr)
```

```python
fig = plt.figure(figsize=(12,8))
plt.scatter(A[:, 0], A[:, 1])
plt.plot(x, f(x, popt[0], popt[1]), color='r', lw=4)
plt.xlabel('X')
plt.ylabel('Y')
plt.xlim([0,10])
plt.ylim([0,10])
```

# Fitting une fonction non lineaire

```python
def another_f(x, a, b, c):
    return a*x**2+b*x+c
```

```python
popt, pcov = curve_fit(another_f, A[:, 0], A[:,1])
perr = np.sqrt(np.diag(pcov))
print (popt, '+/-', perr)
```

```python
fig = plt.figure(figsize=(12,8))
plt.scatter(A[:, 0], A[:, 1], label='data')
plt.plot(x, another_f(x, *popt), color='r', lw=4, label='fit')
plt.xlabel('X')
plt.ylabel('Y')
plt.xlim([0,10])
plt.ylim([0,10])
plt.legend(loc='upper left')
```

## output

```python
def list_files(dir_path):
    fileNames = []
    for f in os.listdir(dir_path):
        if os.path.isfile(dir_path+f):
            fileNames += [dir_path+f,]
    return sorted(fileNames)

def list_dir(dir_path):
    dirNames = []
    for f in os.listdir(dir_path):
        if os.path.isdir(dir_path+f):
            dirNames += [f,]
    return sorted(dirNames)    
```
