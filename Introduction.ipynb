# Example 1a: Générer une sequence de P points consitent de N groupes centrés autour de $(x_i, y_i)$ ou $i$ identifie chaque groupe avec une variance gaussienne spaciale $\sigma_i$.

```python
# definire une nouvelle fonction: "def nom_de_la_fonction(arguments_positionel, argument_nommés=default):"
## les classes des arguments non pas a etre definis explictement. Un usage incorrect result en une erreur.

def some_sequence(P, positions, jitter, return_group=False):
    '''On peut écrire une description qui va ce montrer en tooltip entre triple guilemets'''
    
    # test si les arguments sont consistent
    assert (len(positions)==len(jitter)), 'mismatch'
    
    group = np.random.randint(len(positions), size=(P,))
    A = positions[group] + \
        np.random.normal(0,1,size=(P,)+positions.shape[1:]) \
        * jitter[group].reshape((P,)+(1,)*len(positions.shape[1:]))

    if return_group:
        return A, group
    else:
        return A
```

```python
A = some_sequence(2000, np.random.normal(0,1,size=(4)), np.ones(4)*.1)
```

```python
fig = plt.figure(figsize=(12,4))
_=plt.hist(A, bins=200, range=(-3,3))
plt.xlabel('X')
plt.ylabel('Count')
plt.xlim([-3,3])
```

```python
def list_files(dir_path):
    fileNames = []
    for f in os.listdir(dir_path):
        if os.path.isfile(dir_path+f):
            fileNames += [f,]
    return sorted(fileNames)

def list_dir(dir_path):
    dirNames = []
    for f in os.listdir(dir_path):
        if os.path.isdir(dir_path+f):
            dirNames += [f,]
    return sorted(dirNames)    
```

## Un example d'inheritance

```python
class subdivision_1d(object):
    def __init__(self, n_div=1, dtype=np.float32):
        self.length = n_div
        self.dtype = dtype
        
    def __call__(self):
        '''	returns a list of point positions '''
        return 'base class'
    
class logspace(subdivision_1d):    
    def __init__(self, n_div, dtype=np.float32, **kwargs):
        super(logspace, self).__init__(n_div, dtype=np.float32, **kwargs)
               
    def __call__(self, start, stop):    
        if self.length <= 1:
            return [start]
        lstart = np.log10(start)
        lstop = np.log10(stop)
        return np.logspace(lstart, lstop, self.length)
```

```python
print (subdivision_1d(10)())
```

```python
y = logspace(10)(.1, 100)

print (y)
```
